## Task 1 

The algorithm finds the longest continuous subarray in an array.

The aproach taken for this task is to firstly reduce the array into a set. This is done to remove duplicates and because lookup time in a set is O(1). There are 2 other functions that look forward and backwards to find numbers that would make a run (continuous sub array). 

The time complexity is at most O(N) and space complexity at most O(N).

This code looks in either direction of any number in the set then removes it as it is found. This finds the lenght of every continuos sub array. Only the largest is remembered at any one time. After all elements are removed the largest runs start and end point is returned. 

This program can take an input of a list, array, and set, these are the possible inputs specified in the task description

```python

# Checks to see if the next number is in the set
# Look ups in sets are O(1) so this is constant time algorithm
def next_in(array, current):
    if current+1 in array:
        return True
    else:
        return False


    
# Checks to see if the preivious number is in the set
def previous_in(array, current):
    if current-1 in array:
        return True
    else:
        return False        



# Calculates the lenght of a run from a given number in a set. 
def run_length(array, current):
    # Length of the run
    L = 0
    # How far in above the current number the run goes
    above = current
    # How far in below the current number the run goes
    below = current

    # Looks for the next number greater than the current number in the set, and continues to the end of the run
    while next_in(array, above) == True:
        # removes the element so we dont need to look at it again
        array.remove(above+1)
        L +=1
        above +=1
    
    # Looks for the preivious number less than the current number in the set, and continues to the end of the run
    while previous_in(array, below) == True:
        # Removes the element so we dont look at it again
        array.remove(below-1)
        L +=1
        below -=1
        
    # Returns the length of the run and the begining and end of the run 
    return L, [below,above]
    
    
    
# Finds the longest run
def Largest_range(array):
    # Exceptions
    if len(array) == 0:
        return 'No largest range'

    # If there is no longest run ie all individual 1,3,54
    ans = 'No largest range'
    
    # The largest lenght of a run we've seen
    length = 0

    # Converts the array into a set, this removes duplicates
    # Array to set, this takes O(N) time
    array=set(array)
    
    # Goes through array once, removing visited items in the array (set)
    # This takes at most O(N) time
    while len(array) > 0:
        
        # Any value in the set to start looking at each run that is possible from a starting point, doesn't matter which point we start from
        # Removes the value from the set
        current = array.pop()
        
        # Gets the length and the parameters (low and high) of the run
        L, param = run_length(array, current)

        # If the length of the new run is longer than the current longest the answer and longest length are updated
        if L > length:
            length = L  
            ans = param
    # Returns the start and end of the longest run
    return ans
```

## Task 2

There are three players two strikers and a gaolkeeper. For every goal scored the energy of the respective player is decreased by 1 and after every save the goalkeepers energy is decreased by 1. Strikers will be able to score the goal if Prince energy is a factor of their energy otherwise not.
The session ends when Prince's energy reaches 1.
Assume same player can try for goals repeatedly and they both try to increase number of goals in totality. Lohia (striker X) being a junior player is always favoured for penalty kick and thus kicks first.

I have assumed that the initial input will be a string that prints out a sequence as shown in the task description.

This code follows the instructions given by the task and stated above.

The complexity of this task is at worst requires all the energies to be reducred to 1. Since all the functions involved are constant time the maximum complexity is O(XYZ) however if X,Y ever become prime they will never be able to reduce unless equal to Z.

Given 1 <= T <= 100, 1 <= X,Y,Z <= 105.

```python
# Calculates if a is a factor of b
def factor(a,b):
    # Checks that the remainder on devision is zero, and hence a factor
    if (b/a) % 1 == 0:
        return True



# Calculates the winner of a specific game
def Penalty_score(x,y,z):
    # Exceptions
    if z < 1:
        return 'Invalid input'
    if x < 1:
        return 'Invalid input'
    if y < 1:
        return 'Invalid input'
    
    # Used to calculate the score
    X, Y, Z = x, y, z
    
    # Updates the scores based on the energy of the players
    # Stops when the goalkepers energy is less than one
    while z > 1:
        # Player X goes first 
        # If z is a factor of x, x has scored
        if factor(z,x) == True:
            x -= 1
        # If z is a factor of x, x has scored
        elif factor(z,y) == True:
            y -= 1
        # If neither player scores
        else:
            z -= 1
    # When Z=1 returns the goals scored 
    return X-x, Y-y



# Takes the input and finds the score for each of the energies in input
def Penalty_Shooters(string):
    # Splits the string into the individual lines
    string = string.split('\n')
    count = 0
    
    # Exceptions for incorrect input format
    # Tests that the first line only contains one item
    if len(string[0]) != 1:
        print('Input error')
        return 
    # Tests that T = the number of lines
    if len(string)-1 != int(string[0]):
        print('T not equal to length')
        return 
    # Counter for exceptions
    count = 0
    # Takes each line individually
    for i in string:
        
        # Splits the line into the numbers eg, '1,2,3' becomes [1,2,3]
        i = list(i.split(','))        

        # If there are three energies for players X, Y, Z
        if len(i) == 3:
            # Catches any unforseen exceptions
            try:
                # Each working input adds to the counter so bad input lines can be explained
                count+=1
                
                # Calculates the winner of each of the lines containing the three energies
                x, y = Penalty_score(int(i[0]),int(i[1]),int(i[2]))
                print(x,y)
            except:
                print('Invalid input : line ', count)
                return
```
